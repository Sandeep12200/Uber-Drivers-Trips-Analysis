## Q19. Plot a bar graph of Purpose vs Miles(Distance). What can you infer from the plot(2 +2 points)
#### Note: Use the original dataframe without dropping the 'NA' values.
#### Hint:You have to plot total/sum miles per purpose

```python
total_miles_purpose = ud.groupby('PURPOSE*')['MILES*'].sum()
purpose = total_miles_purpose.index
miles = total_miles_purpose.values

colors = ['#FF6347', '#4682B4', '#32CD32', '#FFD700', '#8A2BE2', '#FF4500', '#2E8B57']

plt.bar(purpose, miles, color=colors, edgecolor='black')

plt.xlabel("Trip Purpose")
plt.ylabel("Total Miles")
plt.title("Total Miles by Trip Purpose")
plt.xticks(rotation=90)

plt.show()

# Observation : "Meeting" has the highest total miles traveled. "Customer Visit" is the second most frequent purpose.
#               "Meal/Entertain" and "Temporary Site" also contribute significantly.
#               Some purposes like "Airport/Travel" and "Moving" have very low total miles.
```

## Q21. Generate a plot showing count of trips vs category of trips. What can you infer from the plot (2 +2 points)
#### Note: Use the original dataframe without dropping the 'NA' values.

```python
trip_count_per_category = ud.groupby('CATEGORY*')['CATEGORY*'].count()
count= trip_count_per_category.values
category= trip_count_per_category.index

plt.pie(count, labels= category, autopct= '%1.1f%%', startangle=45, explode=[0,0.2], wedgeprops={'edgecolor': 'black'})
plt.title("Trip Category VS Count of Trips")
plt.legend(fontsize=8)

plt.show()

# Observation : Most (93.3%) of the trips belong to the Business category. This means that the users primarily uses Uber for professional use.
#               Minimal Personal Travel, Only 6.7% of the journeys are of type Personal.
```

## Q22. What percentage of Miles were clocked under Business Category and what percentage of Miles were clocked under Personal Category ? (3 points)

### Note:Use the original dataframe without dropping the 'NA' values.


```python
miles_category = ud.groupby('CATEGORY*')['MILES*'].sum()
total_miles = miles_category.sum()

business_miles = round((miles_category.get('Business')/total_miles)*100 , 2)
personal_miles = round((miles_category.get('Personal')/total_miles)*100 , 2)

print("Percentage of miles under Business Category:",business_miles,"%")
print("Percentage of miles under Personal Category:",personal_miles,"%")
```
